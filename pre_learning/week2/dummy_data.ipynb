### 【問題1】乱数の作成
```
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
```

````python
import numpy as np
 
np.random.seed(0)
mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
values = np.random.multivariate_normal(mu, sigma, 500)
````

### 【問題2】散布図による可視化
```
問題1で作成したデータ点を散布図により可視化してください。
```

````python
from matplotlib import pyplot as plt
import seaborn as sns

plt.scatter(x=values[:,0], y=values[:,1])
sns.jointplot(x=values[:,0], y=values[:,1])

plt.show()
````

### 【問題3】ヒストグラムによる可視化
```
問題1で作成したデータをヒストグラムにより可視化してください。
```

````python
plt.xlim(-6,3)
plt.hist(x=values[:,0], bins=50)
plt.show()
plt.xlim(-6,3)
plt.hist(x=values[:,1], bins=50)
plt.show()
````

### 【問題4】データの追加
```
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
```

````python
np.random.seed(0)
values2 = np.random.multivariate_normal([0, -3], sigma, 500)
plt.scatter(x=values[:,0], y=values[:,1])
plt.scatter(x=values2[:,0], y=values2[:,1])
plt.show()
````

### 【問題5】データの結合
```
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。
```

````python
join_values = np.concatenate([values, values2])
plt.scatter(x=join_values[:,0], y=join_values[:,1])
plt.show()
````
