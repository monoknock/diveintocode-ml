### 【問題1】データの取得
```
以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。
データが読み込めたら、説明変数を下のようなPandasのDataFrame型変数Xに格納してください。列の名前も以下を参考にしてください。
```

````python
from sklearn.datasets import load_iris
import pandas as pd

iris_dataset = load_iris()

x = pd.DataFrame(iris_dataset.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])
y = pd.Series(iris_dataset.target, name="y")
````

### 【問題2】データの結合
```
問題1で準備したX,yを結合させたDataFrame型変数dfを作成してください。結合することで扱いやすくなります。これ以降はこのdfを使っていきます。
```

````python
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 10)

df = pd.concat([x, y], axis=1)
print(df)
````

### 【問題6】図の作成
```
データは可視化して確認することが大切です。次の3つを行ってください。作図のライブラリはMatplotlibの他seabornもよく使われます。

ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。
```

````python
import numpy as np
import matplotlib.pyplot as plt

# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
plt.pie(df["y"].value_counts(), labels=[0, 1, 2], autopct="%1.1f%%")
plt.title("y_label sample num")

# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
df.plot.box()
plt.grid()
plt.show()

# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する
plt.violinplot(df, showmedians=True)
plt.xticks([1, 2, 3, 4, 5], df.columns.values)
plt.show()

# 箱ひげ図とバイオリン図 の違い
# ▼ 箱ひげ図
#    データの要約統計量を示しつつ、同時に分布を表せるようなグラフ
# ▼ バイオリン図
#    箱ひげ図の分布がより直感的に表示される
#    ただし錯覚を引き起こす要因を秘めているので注意して扱う必要がある
````

### 【問題7】特徴量間の関係の確認
```
特徴量間の関係を見ていきます。次の4つを行ってください。

４つの特徴量から２つ選ぶようなすべての組み合わせ（６つ）の散布図を種類ごとに色分けして作る。
以下の図はそのうち一つの例です。
```

````python
# ４つの特徴量から２つ選ぶようなすべての組み合わせ（６つ）の散布図を種類ごとに色分けして作る。
# "sepal_length", "sepal_width", "petal_length", "petal_width"
from matplotlib import pyplot as plt

fig, axes = plt.subplots(2,3, figsize=(10,7), dpi=70)

axes[0][0].scatter(x=df[df["y"] == 0].loc[:,  ["sepal_length"]], y=df[df["y"] == 0].loc[:,  ["sepal_width"]], c="red")
axes[0][0].scatter(x=df[df["y"] == 1].loc[:,  ["sepal_length"]], y=df[df["y"] == 1].loc[:,  ["sepal_width"]], c="blue")
axes[0][0].scatter(x=df[df["y"] == 2].loc[:,  ["sepal_length"]], y=df[df["y"] == 2].loc[:,  ["sepal_width"]], c="green")

axes[0][1].scatter(x=df[df["y"] == 0].loc[:,  ["sepal_length"]], y=df[df["y"] == 0].loc[:,  ["petal_length"]], c="red")
axes[0][1].scatter(x=df[df["y"] == 1].loc[:,  ["sepal_length"]], y=df[df["y"] == 1].loc[:,  ["petal_length"]], c="blue")
axes[0][1].scatter(x=df[df["y"] == 2].loc[:,  ["sepal_length"]], y=df[df["y"] == 2].loc[:,  ["petal_length"]], c="green")

axes[0][2].scatter(x=df[df["y"] == 0].loc[:,  ["sepal_length"]], y=df[df["y"] == 0].loc[:,  ["petal_width"]], c="red")
axes[0][2].scatter(x=df[df["y"] == 1].loc[:,  ["sepal_length"]], y=df[df["y"] == 1].loc[:,  ["petal_width"]], c="blue")
axes[0][2].scatter(x=df[df["y"] == 2].loc[:,  ["sepal_length"]], y=df[df["y"] == 2].loc[:,  ["petal_width"]], c="green")

axes[1][0].scatter(x=df[df["y"] == 0].loc[:,  ["sepal_width"]], y=df[df["y"] == 0].loc[:,  ["petal_length"]], c="red")
axes[1][0].scatter(x=df[df["y"] == 1].loc[:,  ["sepal_width"]], y=df[df["y"] == 1].loc[:,  ["petal_length"]], c="blue")
axes[1][0].scatter(x=df[df["y"] == 2].loc[:,  ["sepal_width"]], y=df[df["y"] == 2].loc[:,  ["petal_length"]], c="green")

axes[1][1].scatter(x=df[df["y"] == 0].loc[:,  ["sepal_width"]], y=df[df["y"] == 0].loc[:,  ["petal_width"]], c="red")
axes[1][1].scatter(x=df[df["y"] == 1].loc[:,  ["sepal_width"]], y=df[df["y"] == 1].loc[:,  ["petal_width"]], c="blue")
axes[1][1].scatter(x=df[df["y"] == 2].loc[:,  ["sepal_width"]], y=df[df["y"] == 2].loc[:,  ["petal_width"]], c="green")

axes[1][2].scatter(x=df[df["y"] == 0].loc[:,  ["petal_length"]], y=df[df["y"] == 0].loc[:,  ["petal_width"]], c="red")
axes[1][2].scatter(x=df[df["y"] == 1].loc[:,  ["petal_length"]], y=df[df["y"] == 1].loc[:,  ["petal_width"]], c="blue")
axes[1][2].scatter(x=df[df["y"] == 2].loc[:,  ["petal_length"]], y=df[df["y"] == 2].loc[:,  ["petal_width"]], c="green")
plt.show()

# 特徴量すべてを組み合わせを一度に表示させた、散布図行列を作る。
import seaborn as sns
sns.pairplot(df)
plt.show()

# ４つの特徴量に関する相関係数行列を作る
display(df.corr().style.background_gradient(axis=None))

# 相関係数行列のヒートマップを作る(１つ)
sns.heatmap(df)
plt.show()
````
